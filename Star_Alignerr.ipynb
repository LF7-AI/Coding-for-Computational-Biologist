```python
#Step 5: Run Trimmomatic (pair-end mode)---
!trimmomatic PE -threads 2 \
/content/sra_data/SRR15509125/SRR15509125_1.fastq.gz \
/content/sra_data/SRR15509125/SRR15509125_2.fastq.gz \
/content/sra_data/SRR15509125/SRR15509125_1_paired.fq.gz \
/content/sra_data/SRR15509125/SRR15509125_1_unpaired.fq.gz \
/content/sra_data/SRR15509125/SRR15509125_2_paired.fq.gz \
/content/sra_data/SRR15509125/SRR15509125_2_unpaired.fq.gz \
ILLUMINACLIP:/usr/local/miniconda/share/trimmomatic-0.40-0/adapters/TruSeq3-PE.fa:2:30:10 \
LEADING:3 TRAILING:3 SLIDINGWINDOW:4:20 MINLEN:36
```

EXPLAINATION:


PE	= Paired-End mode

-threads 2	= Use 2 CPU threads

First 2 files	= Input read 1 and read 2 FASTQs

Next 4 files	= Output: paired/unpaired reads for R1 and R2

ILLUMINACLIP	= Remove Illumina adapters using adapter file

LEADING:3	= Trim bases < Q3 from the start

TRAILING:3 = Trim bases < Q3 from the end

SLIDINGWINDOW:4:20	= Scan 4-base window; cut when average Q < 20

MINLEN:36	= Drop reads shorter than 36 bases after trimming


THEN,
ILLUMINACLIP:TruSeq3-PE.fa:2:30:10

ILLUMINACLIP	= The Trimmomatic module for adapter clipping.

TruSeq3-PE.fa	= The adapter sequence file.

2	= Seed mismatches:- Maximum number of mismatches allowed in the initial seed alignment between the read and adapter.

30	= Palindrome clip threshold:- How strict to be when detecting adapter dimers (in paired-end data, when read 1 and read 2 overlap and contain adapter sequences).

10	= Simple clip threshold:- 	How strict to be when detecting adapter contamination in a single read.

```python
#Step 7: Build STAR genome index---
!STAR --runMode genomeGenerate \
--genomeDir /content/genome_index \
--genomeFastaFiles /content/genome_index/small_chr22.fa \
--sjdbGTFfile /content/genome_index/small_chr22.gtf \
--runThreadN 4
```

EXPLAINATION:

--sjdbGTFfile /content/genome_index/ :  GTF = Gene Transfer Format file containing gene annotations (exon, intron, transcript locations).

- STAR uses this to extract splice junctions (sjdb = “splice junction database”). This helps STAR learn where exons connect, and improve alignment across exon-exon boundaries inRNA-Seq reads.
