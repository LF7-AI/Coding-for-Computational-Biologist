```python
#Step-2: Import Modules:
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from collections import Counter
import subprocess
```

```python
# Step-3: Define sequences and write to FASTA File
sequences=[
    SeqRecord(Seq("ATTAAGGGCCCCCACGTGCTACCATG"), id= "seq1", description= ""),
    SeqRecord(Seq("AAGGACCCATCACTGTGCTACTCCAGGG"), id= "seq2", description= ""),
    SeqRecord(Seq("ATGCGTACGTACGTACGTAGCTAGCTAGCTAGCTAGC"), id= "seq3", description= "")

]
na_fasta_file = "na_sequences.fasta"
SeqIO.write(sequences, na_fasta_file, "fasta")

with open(na_fasta_file, 'r') as f:
    nucleic_acid_seqs = f.read()
print(nucleic_acid_seqs)
```

```python
#Step-4: Run Clustal Omega alignment for MSA
aligned_na_file= "aligned_na_sequences.fasta"
command= f"clustalo -i {na_fasta_file} -o {aligned_na_file} --outfmt fasta"
subprocess.run(command, shell=True, check=True)

```

```python
#Step 5: Read and display the aligned sequences
from Bio import AlignIO
alignment = AlignIO.read(aligned_na_file, "fasta")
print("Aligned Sequences:")
for record in alignment:
    print(f"{record.id}: {record.seq}")
```
